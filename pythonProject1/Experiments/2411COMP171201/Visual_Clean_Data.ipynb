```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
```

```python
df = pd.read_csv('Maternal Health Risk Data Set.csv')
df.head()
```

```python
bodytemp = df['RiskLevel']
bodytemp.head()
```

```python
plt.hist(bodytemp)
plt.title('Risk level distribution')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
age_scale = df.iloc[:, 0]
# Reshape the Series into a 2D array 
age_scale = age_scale.values.reshape(-1, 1) 
age_scale = scaler.fit_transform(age_scale)
# For plotting, you might want to convert back to a Series:
age_scale = pd.Series(age_scale.flatten())
plt.hist(age_scale)
plt.title('Age distribution')
plt.show()
```

```python
import seaborn as sns
sns.heatmap(df.iloc[:, :-1].corr(), cmap='coolwarm', annot=True, fmt='.3f')
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
X = df.iloc[:, :-1]
X = imputer.fit_transform(X)
print(X)
```
