```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel='rbf', random_state=2004)
classifier.fit(X_train, y_train)
y_pred_test = classifier.predict(X_test)
y_pred_train = classifier.predict(X_train)
print(f'Accuracy on test set: {classifier.score(X_test, y_test)}')
print(f'Accuracy on train set: {classifier.score(X_train, y_train)}')
```

```python
print(np.concatenate((y_pred_test.reshape(len(y_pred_test),1), y_test.reshape(len(y_test),1)),1))
```

```python
print(np.concatenate((y_pred_train.reshape(len(y_pred_train),1), y_train.reshape(len(y_train),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_train, y_pred_train, labels=classifier.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classifier.classes_)
disp.plot()
plt.gca().grid(False)
```

```python
print(np.concatenate((y_pred_train.reshape(len(y_pred_train), 1), y_train.reshape(len(y_train), 1)), 1))
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_test, y_pred_test, labels=classifier.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classifier.classes_)
disp.plot()
plt.gca().grid(False)
```

```python
from joblib import Parallel, delayed
from matplotlib.colors import ListedColormap
# To li d liu  v  th
X_set, y_set = sc.inverse_transform(X_train), y_train
step_size = 0.5 # iu chnh bc nhy  cn bng tc  v  mn
margin = 5 # iu chnh l  gii hn phm vi ca li
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = step_size),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = step_size))

# Hm d on cho mt phn ca li
def predict_chunk(chunk):
  return classifier.predict(sc.transform(chunk))

# Chia li d liu thnh cc phn nh
n_jobs = 29  # S dng tt c cc li CPU
chunks = np.array_split(np.array([X1.ravel(), X2.ravel()]).T, n_jobs)

# Thc hin d on song song trn cc phn
results = Parallel(n_jobs=n_jobs)(delayed(predict_chunk)(chunk) for chunk in chunks)

# Kt hp kt qu
predictions = np.concatenate(results)

# V  th
plt.contourf(X1, X2, predictions.reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('K-NN (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
