```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
X = np.array(df.iloc[:, :-1].values)
y = np.array(df.iloc[:, -1].values)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
from sklearn.preprocessing import StandardScaler
std_X = StandardScaler()

std_X.fit_transform(X)
```

```python
std_X_train = StandardScaler()
X_train = std_X_train.fit_transform(X_train)
X_test = std_X.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=5)
clf.fit(X, y)
clf_train = LogisticRegression(random_state=5)
clf_train.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y.reshape(len(y), 1)), 1))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y, y_pred, labels=clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
disp.plot()
plt.gca().grid(False)
```

```python
clf.score(X, y)
```

```python
y_pred_train = clf_train.predict(X_train)
print(np.concatenate((y_pred_train.reshape(len(y_pred_train), 1), y_train.reshape(len(y_train), 1)), 1))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_train, y_pred_train, labels=clf_train.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf_train.classes_)
disp.plot()
plt.gca().grid(False)
```

```python
clf_train.score(X_train, y_train)
```

```python
clf_train.score(X_test, y_test)
```

```python
y_pred_test = clf_train.predict(X_test)
print(np.concatenate((y_pred_test.reshape(len(X_test), 1), y_test.reshape(len(X_test), 1)), 1))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_test, y_pred_test, labels=clf_train.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf_train.classes_)
disp.plot()
plt.gca().grid(False)
plt.gca().set_facecolor('white')
for text in disp.text_.ravel():
    text.set_color('white')
plt.show()
```

```python
clf_train.score(X_test, y_test)
```

```python
from matplotlib.colors import ListedColormap
X_set, y_set = std_X_train.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, clf_train.predict(std_X_train.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], color = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
