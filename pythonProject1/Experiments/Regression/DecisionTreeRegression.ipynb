```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
```

```python
df = pd.read_csv('Position_Salaries.csv')
X = np.array(df.iloc[:, 1:-1].values)
y = np.array(df.iloc[:, -1].values)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeRegressor
r1_train = DecisionTreeRegressor()
r2_train = DecisionTreeRegressor(max_depth=2)
r3_train = DecisionTreeRegressor(max_depth=5)
r1_train.fit(X_train, y_train)
r2_train.fit(X_train, y_train)
r3_train.fit(X_train, y_train)
```

```python
from sklearn.tree import DecisionTreeRegressor
r1 = DecisionTreeRegressor()
r2 = DecisionTreeRegressor(max_depth=2)
r3 = DecisionTreeRegressor(max_depth=5)
r1.fit(X, y)
r2.fit(X, y)
r3.fit(X, y)
```

```python
plt.scatter(X, y, color='red')
plt.plot(X, r1.predict(X), color='blue')
plt.plot(X, r2.predict(X), color='orange')
plt.plot(X, r3.predict(X), color='purple')
plt.xlabel('PositionLevel')
plt.ylabel('Salary')
plt.title('Decision tree')
plt.show()
```

```python
max_x = max(max(X_train), max(X_test))
X_grid = np.arange(min(X_train), max_x + 0.1, 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
```

```python
plt.scatter(X_test, y_test, color='red')
plt.plot(X_grid, r1_train.predict(X_grid), color='blue')
plt.plot(X_grid, r2_train.predict(X_grid), color='orange')
plt.plot(X_grid, r3_train.predict(X_grid), color='purple')
plt.xlabel('PositionLevel')
plt.ylabel('Salary')
plt.title('Decision tree')
plt.show()
```
