```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
matplotlib.use('TkAgg')
```

```python
data = pd.read_csv('Position_Salaries.csv')
X = np.array(data.iloc[:, 1:2].values)
y = np.array(data.iloc[:, -1].values)
```

```python
print(X)
```

```python
print(y)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor_1 = RandomForestRegressor(n_estimators=10, random_state=0)
regressor_1.fit(X, y)
regressor_2 = RandomForestRegressor(n_estimators=50, random_state=0)
regressor_2.fit(X, y)
regressor_3 = RandomForestRegressor(n_estimators=100, random_state=0)
regressor_3.fit(X, y)
```

```python
regressor_1.predict([[6.5]])
```

```python
X_grid = np.arange(X.min(), X.max(), 0.1)
X_grid = X_grid.reshape(-1, 1)
print(X_grid)
```

```python
plt.title('Random Forest')
plt.scatter(X, y, color='red')
plt.plot(X_grid, regressor_1.predict(X_grid), color='blue')
plt.plot(X_grid, regressor_2.predict(X_grid), color='orange')
plt.plot(X_grid, regressor_3.predict(X_grid), color='yellow')
plt.xlabel('PositionLevel')
plt.ylabel('Salary')
plt.show()
```
